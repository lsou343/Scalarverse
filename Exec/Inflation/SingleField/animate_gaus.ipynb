```python
import yt
import numpy as np
#from yt.mods import *
from yt.frontends.boxlib.api import NyxDataset, AMReXDataset
from yt.data_objects.time_series import DatasetSeries
import matplotlib.pyplot as plt
from matplotlib import rc_context
from matplotlib.animation import FuncAnimation
from IPython import display
```

```python
class AMReXDatasetSeries(DatasetSeries):
    _dataset_cls = AMReXDataset
```

```python
#ts = yt.load("GasSloshingLowRes/sloshing_low_res_hdf5_plt_cnt_*")
path = "plt0????"
ds = AMReXDatasetSeries(path)
#ds = load(path)
```

```python
plot = yt.SlicePlot(ds[0], "z", "KGfdens", center=[0,0,0])
#plot.set_zlim(("boxlib", "KGfdens"), 1e-13, 2e1)
plot.zoom(1)
plot.set_log("KGfdens", False)
plot.show()
```

```python
fig = plot.plots[("boxlib", "KGfdens")].figure
```

```python
def animate(i):
    ts = ds[i]
    plot._switch_ds(ts)
```

```python
animation = FuncAnimation(fig, animate, frames=len(ds))
```

```python
# Override matplotlib's defaults to get a nicer looking font
with rc_context({"mathtext.fontset": "stix"}):
    #animation.save("animation.mp4")
    video = animation.to_html5_video()
    html = display.HTML(video)
    display.display(html)
```
